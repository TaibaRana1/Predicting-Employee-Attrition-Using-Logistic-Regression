****Data Cleaning****

```python
#removing cols which have same vals for all enteries
print(df['EmployeeNumber'].unique())
print(df['EmployeeCount'].unique())
print(df['Over18'].unique())
print(df['StandardHours'].unique())



df = df.drop(['EmployeeCount', 'EmployeeNumber','Over18','StandardHours'], axis=1)
```

```python
for column in df.columns:
  if df[column].dtype in['int64']:
    Q1=df[column].quantile(0.25)
    Q3=df[column].quantile(0.75)
    IQR=Q3-Q1
    LowerBound=Q1-1.5*IQR
    UpperBound=Q3+1.5*IQR

    outliers=df[(df[column]<LowerBound )| (df[column] > UpperBound)]


    if len(outliers)>0:
      print("\n..........................................")
      print(f"Outliers in '{column}':")
      print(f"Number of outliers in '{column}': {len(outliers)}")



print("\nVisualizing the outliers:\n")
#Visualizing outliers only for num cols
for column in df.columns:
  if df[column].dtype in['int64']:
    plt.figure(figsize=(5,2))
    sns.boxplot(x=df[column])
    plt.show()


```

```python
#removing outliers

for column in df.columns:
  if df[column].dtype in['int64']:
    Q1=df[column].quantile(0.25)
    Q3=df[column].quantile(0.75)
    IQR=Q3-Q1
    LowerBound=Q1-1.5*IQR
    UpperBound=Q3+1.5*IQR

    df[column]=df[column].clip(lower=LowerBound, upper=UpperBound)


print("\nVisualizing the outliers after cleaning: \n")
#Visualizing outliers only for num cols
for column in df.columns:
  if df[column].dtype in['int64']:
    plt.figure(figsize=(5,2))
    sns.boxplot(x=df[column])
    plt.show()
```

```python
#removing space and converting to lower case
for column in df.columns:
    if df[column].dtype == 'object' or df[column].dtype.name == 'category':
        df[column] = df[column].astype(str).str.strip().str.lower()
```

****One Hot Encoding****

```python

df['Attrition'] = df['Attrition'].astype(str).str.strip().str.lower()
df['Attrition'] = df['Attrition'].map({'yes': 1, 'no': 0})

df['OverTime'] = df['OverTime'].map({'yes': 1, 'no': 0})
df['Gender'] = df['Gender'].map({'Male': 1, 'Female': 0})

cat_cols = df.select_dtypes(include=['object']).columns
print("Categorical columns to encode:", list(cat_cols))

# One-Hot Encoding
df = pd.get_dummies(df, columns=cat_cols, drop_first=True)

# Converting boolean columns to integers
df = df.astype({col: int for col in df.select_dtypes(include='bool').columns})

print("\nEncoded DataFrame shape:", df.shape)
#print(df.head(5))
```

```python

df['loyalty_ratio'] = df['YearsAtCompany'] / (df['TotalWorkingYears'] + 1)
df['promotion_delay_ratio'] = df['YearsSinceLastPromotion'] / (df['YearsAtCompany'] + 1)
df['normalized_income'] = df['MonthlyIncome'] / (df['JobLevel'] + 1)



df['promotion_delay_ratio'] = df['promotion_delay_ratio'].clip(upper=5)
df['normalized_income'] = df['normalized_income'].clip(upper=20000)
```
